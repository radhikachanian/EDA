#### Checking for NaN values using a heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isna(), yticklabels=False,cbar=False, cmap="viridis");
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df_country.columns
```

```python
df_merged = pd.merge(df, df_country, on = 'Country Code', how = 'inner')
df_merged.head()
```

```python
df_merged.columns
```

```python
df_merged["Country"].value_counts()
```

* From the above cell, we can make an observation that zomato has a huge demand or operational in India, compared to other countries.

```python
#Pie Chart for the country names and its respective values
plt.figure(figsize=(6,6))
plt.pie(x = df_merged["Country"].value_counts().values[:3],
        labels=df_merged["Country"].value_counts().index[:3],
       textprops={'fontsize':7},
       autopct='%1.2f%%')
plt.tight_layout();
```

* OBSERVATION: Majority of the zomato services are operational in India (94.39%), followed by United States (4.73%), and United Kingdom (0.87%).

```python
ratings=df_merged.groupby(["Aggregate rating", "Rating color",
                           "Rating text"])\
                .size().reset_index()\
                .rename(columns={0: "Rating count"})
ratings.head()
```

* OBSERVATION:
1. When rating is between 4.5 - 4.9 --> Excellent 
2. When rating is between 4.0 - 4.4 --> Very Good
3. When rating is between 3.5 - 3.9 --> Good
4. When rating is between 2.5 - 3.4 --> Average
5. When rating is between 2.0 - 2.4 --> Poor

```python
plt.figure(figsize=(10,6), dpi=100)
sns.barplot(x="Aggregate rating", y="Rating count", data=ratings,
           hue="Rating color",
           palette=["blue", "red", "orange", "yellow", "green", "green"])
```

* OBSERVATION:
1. Not rated count is very high
2. Maximum number ratings is between 2.5 - 3.4


```python
sns.countplot(x="Rating color", data=ratings, 
              palette=["blue", "red", "orange", "yellow", "green", "green"])
plt.ylabel("Frequency")
```
