## *2.2. Handling missing values*

```python
#Percentage of missing values 
def percent_missing(df):
    percent_nan = 100 * df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    return percent_nan

result_nan = percent_missing(df)
plt.figure(figsize=(12,7),dpi=200)
sns.barplot(x=result_nan.index, y=result_nan)
plt.xticks(rotation=90);
```

> Here we can observe that Minimum Wage has the maximum number of missing values, followed by Tax Revenue(%), Armed Forces size, Gasoline Price, Population: Labor force participation (%), and Unemployment rate.

Since Minimum Wage has more than 20% of missing values, it would be a better option to replace those missing values by the mean.

```python
def random_nan(df, variable,median): 
    """
    A function to replace NAN values with random sampling
    """
    df[variable+"_median"] = df[variable].fillna(median)
    df[variable+"_random"] = df[variable]
    
    #This will have the random sample to fill the na
    random_sample = df[variable].dropna().sample(
    df[variable].isnull().sum(), random_state=0)
    
    #pandas need to have same index in order to merge the dataset
    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(), variable+"_random"] = random_sample
```

```python
random_nan(df,"Minimum wage", df["Minimum wage"].median)
random_nan(df,"Tax revenue (%)", df["Tax revenue (%)"].median)
random_nan(df,"Armed Forces size", df["Armed Forces size"].median)
random_nan(df,"Gasoline Price", df["Gasoline Price"].median)
random_nan(df,"Population: Labor force participation (%)", df["Population: Labor force participation (%)"].median)
random_nan(df,"Unemployment rate", df["Unemployment rate"].median)
```

```python
fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2,figsize=(12,10))

df["Minimum wage"].plot(kind="kde", ax=ax1)
df["Minimum wage_random"].plot(kind="kde", ax=ax1, color="red")      
lines, labels = ax1.get_legend_handles_labels()
ax1.legend(lines, labels, loc="best")

df["Tax revenue (%)"].plot(kind="kde", ax=ax2)
df["Tax revenue (%)_random"].plot(kind="kde", ax=ax2, color="red")      
lines, labels = ax2.get_legend_handles_labels()
ax2.legend(lines, labels, loc="best")

df["Armed Forces size"].plot(kind="kde", ax=ax3)
df["Armed Forces size_random"].plot(kind="kde", ax=ax3, color="red")      
lines, labels = ax3.get_legend_handles_labels()
ax3.legend(lines, labels, loc="best")

df["Gasoline Price"].plot(kind="kde", ax=ax4)
df["Gasoline Price_random"].plot(kind="kde", ax=ax4, color="red")      
lines, labels = ax4.get_legend_handles_labels()
ax4.legend(lines, labels, loc="best")

df["Population: Labor force participation (%)"].plot(kind="kde", ax=ax5)
df["Population: Labor force participation (%)_random"].plot(kind="kde", ax=ax5, color="red")      
lines, labels = ax5.get_legend_handles_labels()
ax5.legend(lines, labels, loc="best")

df["Unemployment rate"].plot(kind="kde", ax=ax6)
df["Unemployment rate_random"].plot(kind="kde", ax=ax6, color="red")      
lines, labels = ax6.get_legend_handles_labels()
ax6.legend(lines, labels, loc="best")
```

> Here, the distribution is retained, by filling in the missing values

```python
def impute_nan(df, variable):
    """
    Function for filling null values with 0 (indicating missing), assuming they are missing.
    """
    df[variable] = np.where(df[variable].isnull(), 0,df[variable])

    
nan_values_list = ["Total tax rate", "CPI","CPI Change (%)", "Maternal mortality ratio", "Gross tertiary education enrollment (%)", "Life expectancy", "Fertility Rate",
               "Forested Area (%)", "Gross primary education enrollment (%)", "Out of pocket health expenditure", "Physicians per thousand", "Urban_population",
                  "Birth Rate", "Co2-Emissions", "Infant mortality", "Latitude", "Longitude","Agricultural Land( %)"]    
for i in nan_values_list:
    impute_nan(df, i)
```

```python
#Imputting categorical values with "Missing"
def filling_missing(df, variable):
    """
    Function for filling object type with "Missing" 
    """
    df[variable] = np.where(df[variable].isnull(),"Missing", df[variable])
    
filling_missing(df,["Currency-Code", "Largest city"])
```

* ## Top 10 Countries with Highest Population

```python
import matplotlib.ticker as ticker
fig,ax = plt.subplots(figsize=(8,5), dpi=200)
population_count = df.sort_values(by='Population', ascending=False).head(10)
sns.barplot(x="Population", y="Country", data=population_count, ax=ax)
ax.set(xlabel="Population in billions",
      ylabel="Country",
      title="Top 10 Highest Populated Countries")

ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
plt.xticks(rotation=90);

# Annotate the bars with their values
for index, value in enumerate(population_count['Population']):
    ax.text(value + 0.1, index, f'{value:.1f}', va='center', fontsize=10)

plt.show()
```
